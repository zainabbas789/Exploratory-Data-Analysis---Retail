# Exploratory Data Analysis - Retail

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('SampleSuperstore.csv')
```

```python
df.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
catogorical_feature=[ i for i in df.columns if df.dtypes[i]=='object']
```

```python
for j in catogorical_feature:
    print(j)
    print(df[j].unique())
```

```python
df.drop('Postal Code',axis='columns',inplace=True)
```

```python
import seaborn as sns 
sns.pairplot(df)
```

```python
corr=df.corr()
sns.heatmap(corr)
```

```python
df['Ship Mode'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```

```python
df['Segment'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```

```python
df['Category'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```

```python
df.groupby("Ship Mode")['Profit'].sum().sort_values().plot(kind='bar')
plt.title("profit on diff mode")
```

```python
df.groupby('Segment')['Profit'].sum().sort_values().plot(kind='bar')
```

```python
df.groupby('Category')['Profit'].sum().sort_values().plot(kind='bar')
```

```python


df.groupby('Country')['Profit'].sum().sort_values().plot(kind='bar')
```

```python
df.groupby('Region')['Profit'].sum().sort_values().plot(kind='bar')
```

```python
plt.figure(figsize=(15,5))
df.groupby('State')['Profit'].sum().sort_values().plot(kind='bar')
plt.ylabel("Profit")
plt.show()
```

```python
df.head()
```

```python
top_cat_s=df.groupby('Sub-Category')['Sales'].sum().nlargest(n=100)
top_cat_p=df.groupby('Sub-Category')['Profit'].sum().nlargest(n=100)
plt.figure(figsize=(15,5))

plt.style.use('seaborn')
top_cat_s.plot(kind='bar',figsize=(15,5),fontsize=15)
top_cat_p.plot(kind='bar',figsize=(15,5),fontsize=15,color='red')
plt.ylabel('TOTAL SALE and  Profit')
```

```python
plt.figure(figsize=(14,12))
staewise=df.groupby('City')['Profit'].sum().nlargest(n=50)
staewise.plot(kind='barh')
plt.xlabel('Profit')
plt.ylabel('City')
```

```python
plt.figure(figsize=(15,5))
dp=df.groupby('Discount')['Profit'].sum().nlargest(n=50)
dp.plot(kind="barh")
plt.xlabel("Profit")
```

```python
plt.figure(figsize = (6,6))
plt.title('Discount')
plt.pie(df['Discount'].value_counts(), labels=df['Discount'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

```python

sns.catplot(x="Region", y="Profit", kind="bar",hue='Category', data=df)
plt.title('Profits on Segment with different Categories ')
plt.show()
```

```python

sns.catplot(x="Region", y="Profit", kind="bar",hue='Ship Mode', data=df)
plt.title('Profits on Segment with different Categories ')
plt.show()
```

```python
sns.catplot(x="Segment", y="Profit", kind="bar",hue='Category', data=df)
plt.title('Profits on Segment with different Categories ')
plt.show()
```
